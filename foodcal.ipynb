```python
import zipfile
import os

zip_path = '/content/drive/MyDrive/yapay zeka proje/Food Dataset.v1-v1.yolov8.zip'
extraction_path = '/content/food_dataset'

# Create the extraction directory if it doesn't exist
os.makedirs(extraction_path, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_path)

print(f"Dataset unzipped to: {extraction_path}")
```

```python
print("Contents of the unzipped dataset:")
for root, dirs, files in os.walk('/content/food_dataset'):
    level = root.replace('/content/food_dataset', '').count(os.sep)
    indent = ' ' * 4 * (level)
    print(f'{indent}{os.path.basename(root)}/')
    subindent = ' ' * 4 * (level + 1)
    for f in files:
        print(f'{subindent}{f}')
```
